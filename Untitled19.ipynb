```python
# ==============================
# WEEK 7: STATISTICAL ANALYSIS
# ==============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

# ------------------------------
# Load Data
# ------------------------------

# sales = pd.read_csv("sales_data.csv") # Original line
sales = pd.read_csv("/content/sample_data/california_housing_train.csv") # Placeholder fix
# churn = pd.read_csv("customer_churn.csv") # This file is also missing, needs to be uploaded or path corrected

print("Sales Data Preview:")
print(sales.head())

# ------------------------------
# DAY 1: DESCRIPTIVE STATISTICS
# ------------------------------

print("\nDescriptive Statistics:")
desc_stats = sales.describe()
print(desc_stats)

mean_sales = sales['median_house_value'].mean()
median_sales = sales['median_house_value'].median()
mode_sales = sales['median_house_value'].mode()[0]
std_sales = sales['median_house_value'].std()

print(f"\nMean Sales: {mean_sales}")
print(f"Median Sales: {median_sales}")
print(f"Mode Sales: {mode_sales}")
print(f"Standard Deviation: {std_sales}")

# ------------------------------
# DAY 2: DATA DISTRIBUTION
# ------------------------------

plt.figure()
sns.histplot(sales['median_house_value'], kde=True)
plt.title("Median House Value Distribution")
plt.show()

# Normality Test
shapiro_test = stats.shapiro(sales['median_house_value'])
print("\nShapiro Test:", shapiro_test)

# ------------------------------
# DAY 3: CORRELATION ANALYSIS
# ------------------------------

correlation = sales.corr()
print("\nCorrelation Matrix:")
print(correlation)

plt.figure()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# ------------------------------
# DAY 4: HYPOTHESIS TESTING
# ------------------------------

# Hypothesis Test 1: Marketing Spend vs Sales
corr_coef, p_value = stats.pearsonr(sales['median_income'], sales['median_house_value'])

print("\nHypothesis Test 1: Median Income vs Median House Value")
print("Correlation:", corr_coef)
print("P-value:", p_value)

# Hypothesis Test 2: T-Test (High vs Low Marketing)
high_marketing = sales[sales['median_income'] > sales['median_income'].median()]
low_marketing = sales[sales['median_income'] <= sales['median_income'].median()]

t_stat, p_val = stats.ttest_ind(high_marketing['median_house_value'], low_marketing['median_house_value'])

print("\nHypothesis Test 2 (T-Test): High vs Low Median Income on Median House Value")
print("T-stat:", t_stat)
print("P-value:", p_val)

# Hypothesis Test 3: ANOVA (Region vs Sales if region column exists)
if 'Region' in sales.columns:
    groups = [group["median_house_value"].values for name, group in sales.groupby("Region")]
    f_stat, anova_p = stats.f_oneway(*groups)
    print("\nANOVA Test: Region vs Median House Value")
    print("F-stat:", f_stat)
    print("P-value:", anova_p)

# ------------------------------
# DAY 5: CONFIDENCE INTERVAL
# ------------------------------

confidence = 0.95
n = len(sales['median_house_value'])
mean = mean_sales
std_err = stats.sem(sales['median_house_value'])

margin_error = std_err * stats.t.ppf((1 + confidence) / 2., n-1)
ci_lower = mean - margin_error
ci_upper = mean + margin_error

print("\n95% Confidence Interval for Median House Value:")
print(f"{ci_lower} to {ci_upper}")

# ------------------------------
# DAY 6: REGRESSION ANALYSIS
# ------------------------------

X = sales['median_income']
y = sales['median_house_value']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print("\nRegression Summary: Median Income vs Median House Value")
print(model.summary())

# ------------------------------
# DAY 7: BUSINESS INSIGHTS
# ------------------------------

print("\nBUSINESS INSIGHTS:")
if p_value < 0.05:
    print("Median Income significantly affects Median House Value.")
else:
    print("Median Income does NOT significantly affect Median House Value.")

print("R-squared:", model.rsquared)
```
